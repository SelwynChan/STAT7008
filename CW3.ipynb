```python
#find all occurrences of 'Consolidated Statements of Cash Flows'
title_pattern = re.compile(r'Consolidated Statements of Cash Flows\n.*', flags = re.M|re.S)
```

```python
for i in range(len(extracted_text)):
    x = title_pattern.findall(extracted_text[i])
    if not x == []:
        text = x[0]
```

```python
#get each line of the statement
line_pattern = re.compile(r'[A-za-z,\—\-\s]+ \.[ \.]+[\n0-9,\—\$\)\(]+')
lines = line_pattern.findall(text)
lines
```

```python
name_pattern = re.compile(r'[A-Z][a-z,\—\-\s]+')
names = [name_pattern.findall(line) for line in lines]
names
```

```python
#get figures

#remove comma from figures
lines = [line.replace(',','') for line in lines]

fig_pattern = re.compile(r'[0-9\)\(\—]+')
figs = [fig_pattern.findall(line) for line in lines]
figs = [fig for fig in figs if not fig == []]
figs
```
